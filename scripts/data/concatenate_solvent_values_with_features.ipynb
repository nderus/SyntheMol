```python
import numpy as np
import pandas as pd
```

```python
#Loading files that are the output of the chemfunc save_fingerprints command. Have separate RDKit and Morgan features .npz files
data_rdkit_abs = np.load('chemfluor_abs_rdkit_train_features.npz')
data_morgan_abs = np.load('chemfluor_abs_morgan_train_features.npz')

data_rdkit_emi = np.load('chemfluor_emi_rdkit_train_features.npz')
data_morgan_emi = np.load('chemfluor_emi_morgan_train_features.npz')

data_rdkit_plqy = np.load('chemfluor_plqy_rdkit_train_features.npz')
data_morgan_plqy = np.load('chemfluor_plqy_morgan_train_features.npz')
```

```python
#PLQY: 3055
#Abs: 4202
#Emission: 433

data_morgan_abs['features'].shape
```

```python
#Extracting the SP, SdP, SA, SB values from the three separate train dataset files
abs_smiles = pd.read_csv("chemfluor_abs_train_solvents.csv")
abs_solvents = abs_smiles[['SP', 'SdP', 'SA', 'SB']].values
emi_smiles = pd.read_csv("chemfluor_emi_train_solvents.csv")
emi_solvents = emi_smiles[['SP', 'SdP', 'SA', 'SB']].values
plqy_smiles = pd.read_csv("chemfluor_plqy_train_solvents.csv")
plqy_solvents = plqy_smiles[['SP', 'SdP', 'SA', 'SB']].values
```

```python
#Combining features in .npz file (_r for RDKit features, _m for Morgan features) with the solvent values.
abs_save_r = np.concatenate((data_rdkit_abs['features'], abs_solvents), axis=1)
abs_save_m = np.concatenate((data_morgan_abs['features'], abs_solvents), axis=1)

em_save_r = np.concatenate((data_rdkit_emi['features'], emi_solvents), axis=1)
em_save_m = np.concatenate((data_morgan_emi['features'], emi_solvents), axis=1)

plqy_save_r = np.concatenate((data_rdkit_plqy['features'], plqy_solvents), axis=1)
plqy_save_m = np.concatenate((data_morgan_plqy['features'], plqy_solvents), axis=1)
```

```python
#Saving the files as new .npz files to input into the training commands
np.savez('chemfluor_abs_rdkit_train_features_solvent.npz', features=abs_save_r)
np.savez('chemfluor_abs_morgan_train_features_solvent.npz', features=abs_save_m)

np.savez('chemfluor_emi_rdkit_train_features_solvent.npz', features=em_save_r)
np.savez('chemfluor_emi_morgan_train_features_solvent.npz', features=em_save_m)

np.savez('chemfluor_plqy_rdkit_train_features_solvent.npz', features=plqy_save_r)
np.savez('chemfluor_plqy_morgan_train_features_solvent.npz', features=plqy_save_m)
```
